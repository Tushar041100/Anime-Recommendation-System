```python
animelist.drop_duplicates(inplace=True)
```

```python
animelist.drop(columns=['watching_status','watched_episodes'], inplace=True)
```

```python
animelist.head(1)
```

```python
animelist['user_id'].nunique()
```

```python
avg_user_ratings = animelist['user_id'].value_counts().mean().__ceil__()
```

```python
# Filter users with ratings greater than avg_user_ratings
user_counts = animelist['user_id'].value_counts()
filtered_users = user_counts[user_counts > avg_user_ratings].index
animelist = animelist[animelist['user_id'].isin(filtered_users)]
```

```python
print(animelist.shape)
```

```python
anime = anime.merge(anime_with_synopsis, on='MAL_ID', how='inner')
```

```python
anime.dropna(inplace=True)
```

```python
anime.shape
```

```python
anime.head(1)
```

```python
anime.replace("Unknown", np.nan)
```

```python
anime.head(1)
```

```python
ratings = animelist.groupby('anime_id')['rating'].mean().reset_index()
ratings.shape
```

```python
ratings.head(1)
```

```python
anime = anime.rename(columns={'MAL_ID':'anime_id'})
```

```python
anime.head(1)
```

```python
anime_ratings = anime.merge(ratings, on='anime_id', how='inner')
anime_ratings.shape
```

```python
anime_ratings.head(1)
```

```python
anime_ratings.isin(['Unknown']).sum()
anime_ratings.replace("Unknown", np.nan, inplace=True)
```

```python
anime_ratings.isnull().sum()
```

```python
anime_ratings['Name_x'] = anime_ratings.apply(lambda row: row['English name'] if pd.notnull(row['English name']) else row['Name_x'], axis=1)
```

```python
##id name score genre english name Japanese name episodes studios

anime_ratings = anime_ratings[['anime_id', 'Name_x', 'Genres_x', 'Episodes', 'Popularity', 'Members', 'Favorites', 'rating', 'sypnopsis', 'Ranked', 'Score_x']]
```

```python
anime_ratings.head(1)
```

```python
anime_ratings.isin(['Unknown']).sum()
```

```python
popular_anime = anime_ratings[anime_ratings['Score_x'] != 'Unknown'].sort_values('Popularity', ascending=True).head(50)
popular_anime
```

```python
new_score_system = anime_ratings
new_score_system['Score_x'] = new_score_system['Score_x'].replace('Unknown', 0)
new_score_system['rating'] = new_score_system['rating'].replace('Unknown', 0)

new_score_system['Score'] = np.where(new_score_system['Score_x'] != 0,
                                (new_score_system['Score_x'].astype(float) + new_score_system['rating'].astype(float)) / 2,
                                new_score_system['rating'].astype(float))
```

```python
new_score_system = new_score_system.sort_values('Score', ascending=False).head(50)
new_score_system
```

```python
## Reset index
popular_anime.reset_index(drop=True, inplace=True)
new_score_system.reset_index(drop=True, inplace=True)
```

```python
## Popularity wise rating anime
popular_anime
```

```python
## Score wise rating anime
new_score_system.drop(columns=['Score_x', 'rating', 'Popularity', 'Episodes', 'Favorites', 'Members', 'Ranked'], inplace=True)

new_score_system = new_score_system.rename(columns={'Name_x':'Name', 'Genres_x':'Genres', 'sypnopsis':'Synopsis', 'Score':'Rating', 'anime_id':'Anime ID'})
```

```python
new_score_system['Rating'] = new_score_system['Rating'].round(2)
new_score_system
```

```python
import pickle
pickle.dump(new_score_system, open('popular_anime.pkl', 'wb'))
```
